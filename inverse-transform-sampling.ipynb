# Inverse Transform Sampling with Python

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import expon
```

## Generate an exponential distribution

```python
# Create exponential distribution wih lambda
LAMBDA = 1.5
```

```python
# Plot the distribution
x = np.linspace(0, 5, 1000)
y = LAMBDA * np.exp(-LAMBDA * x)
plt.plot(x, y, color='orange')
plt.title('Exponential distribution')
plt.show()
```

## Perform inverse sampling

```python
# Sample from uniform distribution, and transforming it into exponential distribution
SAMPLES = 1000
z = np.random.uniform(low=0, high=1, size=SAMPLES)
y = - (1 / LAMBDA) * np.log(1 - z)
```

```python
# Create axis
fig, ax = plt.subplots(1, 1)

# Plot the distribution
sns.histplot(y, stat='density', ax=ax)
x = np.linspace(0, 5, 1000)
pdf_values = LAMBDA * np.exp(-LAMBDA * x)
ax.plot(x, pdf_values, color='green', linewidth=4)
plt.title('Exponential distribution')
plt.savefig('figures/inverse-transform-sampling.png')
plt.show()
```
